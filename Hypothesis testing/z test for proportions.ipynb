## Z-test for Porportions 

This is used to test whether a sample poportion p differes from a known or hypothesized population $p_{0}$. This test is particularly useful when dealing with categorical data, such as success/failure and determining whether the observed proportion is significantly different from a given proportion. 

### One-Sample z-test

#### Hypothesis 
- Null hypothesis: $H_{o}: ~p = p_{o}$
- Alternate hypothesis two tailed: $H_{1}: p \neq p_{o}$
- Alternate hypothesis one tailed: $H_{1}: p>p ~or~ p<p_{o}$

#### Assumptions: 
- The sample should large enough i.e. $np_{o}\geq5$ and $n(1-p_{o})\geq5$
- The sample is randomly selected and is independenet. 

#### Statistic 

$$z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1 - p_0)}{n}}}
$$

where 
- $\hat{p}$ is the sample proportion. 
- $p_{o}$ is the hypothesied population proportion.
- $n$ is the sample size.

#### Determine the p-value and rejecting or not rejecting the hypothesis
- Using the z-value we wind the p value if the p-value is less than $\alpha$, we reject the hypothesis, or do not reject the null hypothesis. 

- We can also use the test statistic. Let z be a test statitic value the $|z|>1.96$ we reject the null hypothesis at a 5% significance level. That is, when the test statistic lies outside the 95% confidence interval, we reject the null hypothesis. If it lies within the interval, we do not reject the null hypothesis.

### Two sample-proportion test. 

#### State the Hypotheses: 

- Null hypothesis $H_{o}: ~p_{1} =p_{2}$ ,The population proportions are equal.
- Alternate hypothesis: Two talied $H_{1}: ~p_{1} \neq p_{2}$
- Alternate hypothesis: One sided $H_{1}: ~p_{1} > p_{2} ~or~ ~p_{1} < p_{2}$

#### Caluclate the pooled proportion 

$$\hat{p} = \frac{x_{1}+x_{2}}{n_{1}+n_{2}}$$

where,
- $x_{1},x_{2}$ are the number of successes in each example. 
- $n_{1}$ and $n_{2}$ are the sample sizes 

#### Statistic

$$z = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{\hat{p}(1 - \hat{p})\left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}$$

where,
- $\hat{p_{1}} = \frac{x_{1}}{n_{1}}$ is the sample proportion from group 1.
- $\hat{p_{2}} = \frac{x_{2}}{n_{2}}$ is the sample proportion from group 2.
= $\hat{p}$ is the pooled proportion 

#### Decision 

- For a two-tailed test, the critical values for a significance level $\alpha$
(take $\alpha = 0.05$) correspond to a $z_{critical}$ value of $\pm 1.96$
- One can either compare the test statistic z to the critical value or calculate the p-value from the standard normal distribution.
