## Z-Score

The Z-test is a statistical test used to determine whether there is a significant difference between sample and population means (or between two sample means) when the population variance is known and the sample size is sufficiently large (typically n>30) It’s based on the normal distribution (Z-distribution).

Note: Population mean and variance is known here.

#### Test Statistic

The test statistic in a Z-test is the Z-score, which tells you how many standard deviations a data point (sample mean) is from the population mean.

1. For comparing a sample mean to population mean 

$$Z={\frac{\bar{X}-\mu}{\frac{\sigma}{\sqrt{n}}}}$$

$\begin{array}{r l}{\bullet}&{{}\bar{X}=\,{\mathsf{Sample ~ mean}}}\\
 {\bullet}&{{}\mu=\mathrm{poplation~standard~deviatior}}\\ {\bullet}&
 {{}\sigma={\mathsf{p o p u l a t i o n~s t a n d a r d~d e v i a t i o r}}}\\
  {\bullet}&{{}n={\mathsf{s a m p l e~s i z e}}}\end{array}$

2. For comparing two sample means 

$$Z={\frac{\left(\bar{X_{1}}-\bar{X_{2}}\right)}{\sqrt{\frac{\sigma_{1}^{2}}{n_{1}}}+{\frac{\sigma_{2}^{2}}{n_{2}}}}}$$

$\bullet ~~~~ \bar{X_{1}},\bar{X_{2}} ~ \mathsf{are ~ sample ~means ~of ~two ~samples}\\$
$\bullet ~~~~ \sigma_{1}, \sigma_{2}  ~ \mathsf{are ~the ~standard ~deviations ~for ~the ~two ~samples}\\$ 
$\bullet ~~~~ n_{1} ,n_{2} ~ \mathsf{are ~the ~sample ~sizes} $


#### Types of Z-tests
- One-sample Z-test: Compares the mean of a sample to a known population mean.
- Two-sample Z-test: Compares the means of two independent samples.
- Proportion Z-test: Compares sample proportions to population proportions or between two sample proportions.

#### Assumptions for Z-test

- The sample data should be drawn from a normally distributed population, or the sample size should be large enough for the Central Limit Theorem to apply.
- Population variance or standard deviation must be known.
- Random sampling.

#### Example 
Problem: A factory claims that the average lifetime of its light bulbs is 1000 hours. A random sample of 50 bulbs has a mean lifetime of 980 hours and a standard deviation of 40 hours. Is there evidence at the 0.05 significance level to doubt the factory's claim?

Hypothesis Formulation: There is no significant diffence between the sample and population means 

- Null Hypothesis $H_{o}:  \mu= 1000 \mathsf{(no ~difference ~from ~the ~claimed ~lifetime)}$
- Alternative hypothesis $H_{1}:  \mu \neq 1000 \mathsf{(there ~is ~some ~difference)}$

Given,
- Sample mean $\bar = 980$
- Population mean $\mu = 1000$
- Population standard deviation $\sigma = 40$
- Sample size $n = 50$
- Signifiance level $\alpha = 0.05$

Z-test statistic 
$$Z={\frac{980-1000}{{\frac{40}{\sqrt{50}}}}}={\frac{-20}{5.657}}=-3.54$$

Decision: Using the standard normal distribution table, the critical Z-value for a two-tailed test at 
$\alpha=0.05$ is approximately ±1.96. Since −3.54 is outside this range, we reject the null hypothesis.

Conclusion: There is significant evidence to suggest that the factory's claim about the lifetime of its bulbs may be incorrect.
